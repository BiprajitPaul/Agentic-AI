```python
from pydantic import BaseModel, Field
```

```python
from dataclasses import dataclass
@dataclass # This is a decorator that is used to create a dataclass. A dataclass is a simple way to define a class that has some attributes. It automatically generates some special methods for the class, like __init__, __repr__, and __eq__, based on the attributes that are defined in the class.
class Person:
    name: str
    age: int
    city: str
person = Person(name="Tuhin", age=23, city="Kumarghat")
print(person)
```

This is the difference between a dataclass and a pydantic model. A dataclass is a simple way to define a class that has some attributes, while a pydantic model is a more powerful way to define a class that can also validate the data that is passed to it. In this example, we can see that the pydantic model will raise an error if we try to create an instance with invalid data (like age being a string instead of an integer), while the dataclass will not raise any error and will just create the instance with the invalid data.

```python
class Person(BaseModel):
    name: str
    age: int
    city: str

person = Person(name="Tuhin", age=23, city="Kumarghat")
print(person)
```

```python
person1 = Person(name="Tuhin", age=23, city=12)
print(person1)
```

```python
# Model with Optinal Field
from typing import Optional

class Employee(BaseModel):
    id: int
    name: str
    department: str
    salary: Optional[float] = None ## This means that the salary field is optional and can be set to None if it is not provided when creating an instance of the Employee model.

    is_active: bool = True ## This means that the is_active field is a boolean that defaults to True if it is not provided when creating an instance of the Employee model.
```

```python
print( Employee(id=1, name="Alice", department="HR") ) ## This will create an instance of the Employee model with the id, name, and department fields set to the provided values, and the salary field set to None and the is_active field set to True by default.
```

```python
from typing import List
class Classroom(BaseModel):
    name: str
    students: List[str] ## This means that the students field is a list of strings, which can be used to store the names of the students in the classroom.
    capacity: int
```

```python
classroom = Classroom(name="Math 101", students=["Alice", "Bob", "Charlie"], capacity=30) 
print(classroom)
```

```python
classroom2 = Classroom(name="Math 102", students=("Tuhin", "Sagar", "Trisit"), capacity=30) # tuples are also accepted for list fields in pydantic models, so this will work without any issues.
print(classroom2)
```

```python
## Try catch block

try:
    invalid_val = Classroom(name="Math101", students=["Alice", 123], capacity=30) ## This will raise a validation error because the name field is empty, and the students field contains an integer (123) instead of a string.
    print(invalid_val)
except ValueError as e:
    print("Validation error:", e)
```

```python
# Models with nested models
class Address(BaseModel):
    street: str
    city: str
    state: str
    zip_code: int 
class Customer(BaseModel):
    name: str
    email: str
    address: Address ## This means that the address field is a nested model of type Address, which can be used to store the customer's address information in a structured way.
customer = Customer(
    name="John Doe",
    email="john.doe@example.com",
    address=Address(
        street="123 Main St",
        city="Anytown",
        state="CA",
        zip_code="799290"
    )
)
print(customer)
```

```python
# Pydantic Fields : Customization and constraints

class Item(BaseModel):
    name: str  = Field(min_length=2, max_length=50) ## This means that the name field must be a string with a minimum length of 2 characters and a maximum length of 50 characters.

    price:float=Field(gt=0, le=1000) ## This means that the price field must be a float that is greater than 0 and less than or equal to 1000.
    quantity: int = Field(ge=1) ## This means that the quantity field must be an integer that is greater than or equal to 1.
```

```python
item = Item(name="Laptop", price=23, quantity=1)
print(item)
```

# Model Schema 👇

```python
Item.model_json_schema() # This will generate a JSON schema for the Item model, which can be used to validate data against the model's structure and constraints. The generated JSON schema will include information about the fields, their types, and any validation rules that have been defined using the Field function.
```
