```python
from nltk.stem import PorterStemmer
```

```python
stemming = PorterStemmer()
```

```python
words = ["eating", "eat", "eaten", "writing", "writes", "programming", "programs","history","finally","finalized" ]
```

```python
for word in words:
  print(word + "------>"+stemming.stem(word))
```

```python
# regexpStemmer class
from nltk.stem import RegexpStemmer
```

```python
reg_stemmer = RegexpStemmer('ing$|s$|e$|able$', min=4)
```

```python
reg_stemmer.stem("eating")
```

```python
reg_stemmer.stem("history")
```

```python
# Snowball Stemmer

from nltk.stem import SnowballStemmer
```

```python
snowball = SnowballStemmer("english")
```

```python
for word in words :
  print(word + "------>"+snowball.stem(word))
```

```python
stemming.stem("fairly"),stemming.stem("sportingly"), stemming.stem("goes")
```

```python
snowball.stem("fairly"),snowball.stem("sportingly"),snowball.stem("goes")
```

### 1. Porter Stemmer
**Example:** `stemming.stem("fairly")` -> `'fairli'`

- **Advantages:** It is the oldest and most widely used stemming algorithm. It is simple, fast, and provides a good balance between precision and recall for general English text.
- **Disadvantages:** It is quite aggressive and can produce stems that are not actual words (e.g., 'fairli'). It doesn't handle newer vocabulary as well as more modern stemmers.

### 2. Snowball Stemmer
**Example:** `snowball.stem("fairly")` -> `'fair'`

- **Advantages:** Often called the 'Porter2' stemmer, it is more sophisticated and generally more accurate than the original Porter Stemmer. It supports multiple languages (not just English) and handles suffixes like '-ly' more effectively.
- **Disadvantages:** While more accurate, it is slightly more computationally expensive than Porter, though still very fast.

### 3. Regexp Stemmer
**Example:** `reg_stemmer.stem("eating")` -> `'eat'`

- **Advantages:** Provides total control. You can define specific patterns to strip based on your domain-specific needs (e.g., stripping only 'ing' or 'able').
- **Disadvantages:** It is manual and brittle. If you don't account for a specific suffix, it won't be stemmed. It requires the user to write complex regular expressions and lacks the built-in linguistic logic found in Porter or Snowball.

### General Stemming Summary

| Feature | Porter | Snowball | Regexp |
| :--- | :--- | :--- | :--- |
| **Accuracy** | Moderate | High | Custom |
| **Speed** | Fastest | Fast | Fast |
| **Complexity** | Low | Medium | High (to write) |
| **Use Case** | General NLP | Modern NLP | Domain-specific |

**Common Disadvantage for all:** Over-stemming (stripping too much) or Under-stemming (not stripping enough) can occur, leading to loss of semantic meaning.
