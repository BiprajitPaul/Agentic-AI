```python
paragraph = """ Sardar Vallabhbhai Patel, widely known as the “Iron Man of India,” was one of the most influential leaders in India’s struggle for independence and in shaping the newly formed nation after 1947. Born on October 31, 1875, in Nadiad, Gujarat, he was a barrister by profession before becoming deeply involved in the freedom movement under the leadership of Mahatma Gandhi. Patel played a crucial role in organizing peasants and leading movements such as the Bardoli Satyagraha, which earned him the title “Sardar,” meaning leader. After India gained independence, he served as the first Deputy Prime Minister and Home Minister, where his most remarkable contribution was the political integration of over 560 princely states into the Indian Union, a task that required strong negotiation skills, strategic thinking, and unwavering determination. Without his efforts, India might have remained fragmented into numerous independent regions.
 Patel was known for his firm leadership, practical decision-making, and dedication to national unity. He believed that a strong and united India was essential for progress and stability, and he worked tirelessly to build administrative systems that could support the country’s growth. His vision led to the establishment of modern civil services, which continue to play a vital role in governance today. Despite his immense contributions, he maintained a simple lifestyle and remained deeply connected to the common people. In recognition of his legacy, the Statue of Unity, the world’s tallest statue, was built in his honor, symbolizing his enduring impact on the nation. His life and work continue to inspire generations, reminding us of the importance of unity, integrity, and strong leadership in building a nation.
 """
```

```python
import nltk
from nltk.stem import SnowballStemmer
from nltk.corpus import stopwords
nltk.download('stopwords')

nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
snowball =  SnowballStemmer('english')
```

```python
stopwords.words("english")
```

```python
sentences = nltk.sent_tokenize(paragraph)
```

```python
type(sentences)
```

```python

#Apply stopwords and filter and then apply stemming

for i in range(len(sentences)):
  words = nltk.word_tokenize(sentences[i])
  words = [snowball.stem(word) for word in words if word not in set(stopwords.words("english"))]
  sentences[i] = " ".join(words)
```

```python
# Wordnet Lemmatizer
from nltk.stem import WordNetLemmatizer
import nltk
nltk.download('wordnet')
lemmatizer = WordNetLemmatizer()
```

```python
# This cell converts each sentence to lowercase, tokenizes it, performs lemmatization on words (verbs), and removes English stopwords.
for i in range(len(sentences)):
  sentences[i] = sentences[i].lower()
  words = nltk.word_tokenize(sentences[i])
  words = [lemmatizer.lemmatize(word, pos="v") for word in words if word not in set(stopwords.words("english"))]
  sentences[i] = " ".join(words)
```

Here's a step-by-step explanation of the code in the cell above:

1.  **`for i in range(len(sentences)):`**: This loop iterates through each sentence in the `sentences` list.

2.  **`sentences[i] = sentences[i].lower()`**: Inside the loop, the current sentence (`sentences[i]`) is converted to all lowercase characters. This helps in standardizing the text, ensuring that words like "The" and "the" are treated as the same word.

3.  **`words = nltk.word_tokenize(sentences[i])`**: The lowercase sentence is then broken down into individual words or tokens using `nltk.word_tokenize()`. This creates a list of words from the sentence.

4.  **`words = [lemmatizer.lemmatize(word, pos="v") for word in words if word not in set(stopwords.words("english"))]`**: This is a list comprehension that performs two main operations:
    *   **Stopword Removal**: `if word not in set(stopwords.words("english"))` checks if each `word` is present in a set of common English stopwords (e.g., "a", "the", "is"). If a word is a stopword, it is excluded from the new `words` list.
    *   **Lemmatization**: For the words that are not stopwords, `lemmatizer.lemmatize(word, pos="v")` is applied. This converts the word to its base or dictionary form (lemma), specifically treating it as a verb (`pos="v"`). For example, "running" becomes "run", "was" becomes "be", etc.

5.  **`sentences[i] = " ".join(words)`**: Finally, the processed words (lemmatized and without stopwords) are joined back together with spaces in between to form a new sentence string. This new sentence replaces the original sentence at the same index `i` in the `sentences` list.

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
import nltk
sentences =  nltk.sent_tokenize(paragraph)

nltk.download('averaged_perceptron_tagger_eng')
```

```python
## We will find out pos tag

for i in range(len(sentences)):
  words = nltk.word_tokenize(sentences[i])

  # Rewriting the stopword removal for clarity
  filtered_words = []
  english_stopwords = set(stopwords.words("english"))
  for word in words:
    if word not in english_stopwords:
      filtered_words.append(word)
  words = filtered_words

  pos_tag = nltk.pos_tag(words)
  print(pos_tag)
```
