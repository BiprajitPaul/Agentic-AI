```python
from typing_extensions import Annotated, TypedDict
import operator


```

```python
from langchain_core.messages import AnyMessage , HumanMessage, AIMessage
```

```python
class ektaState(TypedDict):
    messages: Annotated[list[AnyMessage], operator.add]

# What is does is it defines a state that can hold a list of messages, which can be either from a human or an AI. The operator.add annotation indicates that when we add two states together, it will concatenate their message lists.

# And the ektaState(TypedDict) is a way to define a structured data type that can be used to store and manipulate the conversation history in a more organized way.
```
